 Question 1:  What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
 Ans. . First, the big idea:

Anomaly detection is about finding data points or patterns that don‚Äôt fit the "normal" behavior.

Think of it like spotting a needle in a haystack ‚Äî the needle is different, so it stands out.

It's super useful in fraud detection, cybersecurity, health monitoring, manufacturing, etc.

üëâ Quick check: Can you think of an everyday situation where noticing something ‚Äúweird‚Äù is important? (like in banking, health, or even daily life)

2. Types of anomalies:
a) Point Anomalies

A single data point is very different from the rest.

Example: If most credit card transactions are around ‚Çπ500‚Äì‚Çπ2,000, but suddenly one shows ‚Çπ1,00,000, that‚Äôs a point anomaly.

b) Contextual Anomalies

A data point looks normal in general, but abnormal in a given context.

Example: Spending ‚Çπ5,000 on electricity in summer might be fine, but in winter (when usage is low), the same ‚Çπ5,000 bill is suspicious.

Context matters: time, location, season, etc.

c) Collective Anomalies

A group of data points together is abnormal, even if individual points look normal.

Example: One failed login attempt on your account is fine. But 50 failed logins in 10 minutes = collective anomaly ‚Üí possible hacking attempt.

3. Quick recap in one line each:

Point anomaly ‚Üí one weird data point.

Contextual anomaly ‚Üí looks normal overall, but weird in its situation.

Collective anomaly ‚Üí group of values acting weird together.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
Ans. , let‚Äôs break this down into a compare-and-contrast exercise instead of a wall of text.
We‚Äôll look at Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) along three angles:

How they work (approach)

When to use them (use cases)

Strengths / Weaknesses

1. Isolation Forest (IF)

Approach:

Randomly splits data into trees.

Outliers are easier to isolate (take fewer splits).

Best for:

High-dimensional data (like text, finance, cybersecurity logs).

Large datasets (fast and scalable).

Strengths: Handles high-dimensions, efficient.

Weakness: Doesn‚Äôt capture local density well (can miss anomalies in small clusters).

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Approach:

Groups points into dense clusters.

Points in sparse areas = anomalies (noise).

Best for:

Spatial/geographical data, or data where cluster shapes aren‚Äôt spherical.

When you don‚Äôt know the number of clusters.

Strengths: Detects clusters of any shape, finds outliers naturally.

Weakness: Needs good parameter tuning (eps, minPts), struggles with varying densities.

3. Local Outlier Factor (LOF)

Approach:

Compares local density of a point with its neighbors.

If much sparser than neighbors ‚Üí anomaly.

Best for:

Data where local density matters (e.g., fraud detection: unusual spending compared to peers).

Strengths: Catches contextual anomalies.

Weakness: Sensitive to parameter k (number of neighbors), not great for very large datasets.

üîë Simple Mnemonic:

Isolation Forest ‚Üí "Cuts trees fast" (good for big, high-dimensional data).

DBSCAN ‚Üí "Dense = cluster, Sparse = anomaly" (great for spatial clustering).

LOF ‚Üí "Compare with neighbors" (good for local/contextual anomalies).

Question 3: What are the key components of a Time Series? Explain each with one example.
Ans. Let‚Äôs explore time series components in a way that makes them stick.

A time series is data collected over time (daily stock prices, monthly sales, etc.). To understand it, we usually break it into four key components:

1. Trend

The long-term movement (upward or downward) in the data.

Example: A company‚Äôs sales increasing steadily over 5 years.

2. Seasonality

Regular, repeating patterns due to seasons, months, days, etc.

Example: Ice cream sales peak every summer and dip in winter.

3. Cyclic Component

Long-term fluctuations that are not fixed like seasonality.

Related to economic or business cycles (can last years).

Example: Housing market prices rising during a boom and falling during a recession.

4. Irregular / Random Component

Unexpected variations, caused by unusual events.

Example: Sudden drop in airline travel during the COVID-19 pandemic.

üëâ Quick recap:

Trend = overall direction

Seasonality = fixed repeating pattern

Cycle = long-term ups & downs (not fixed)

Random = unpredictable shocks

Let‚Äôs test your grasp:
Suppose you track daily electricity demand for a city.

In winter, demand is high (heaters).

In summer, demand is also high (ACs).

But during a sudden heatwave, demand spikes abnormally.

Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
Ans. . What does Stationary mean?

A stationary time series is one whose statistical properties don‚Äôt change over time.

Mean = constant

Variance = constant

Autocorrelation (relationship with its past values) = doesn‚Äôt change with time

üìå Example: White noise (random fluctuations around zero) is stationary.
But stock prices usually aren‚Äôt (they trend upward or downward).

2. Why is it important?

Many forecasting models (like ARIMA) assume stationarity.
If your data isn‚Äôt stationary, the model can give misleading results.

3. How to test stationarity?

Visual check:

Plot the series ‚Üí if you see clear trends/seasonality, probably non-stationary.

Summary statistics:

Check mean/variance in different time windows. If they shift, non-stationary.

Formal tests:

ADF (Augmented Dickey-Fuller) test: Null hypothesis = non-stationary.

KPSS test: Null hypothesis = stationary.

3. ARIMA (AutoRegressive Integrated Moving Average)

Structure: Combines AR + MA + differencing (I = Integrated) to handle non-stationary data.

Application: Forecasting non-stationary time series with trend.

4. SARIMA (Seasonal ARIMA)

Structure: Extends ARIMA by adding seasonal components.

Non-seasonal part: (p, d, q)

Seasonal part: (P, D, Q, s) where s = season length.

Application: Time series with seasonality (monthly, quarterly, yearly).

Example: Forecasting ice cream sales with yearly seasonality.

5. SARIMAX (Seasonal ARIMA with eXogenous variables)

Structure: SARIMA + external predictors (X).

Exogenous = outside factors that influence the series.

Application: When external features help improve predictions.

Example: Forecasting energy demand using SARIMAX with temperature as an external variable.

üö¶ Quick summary:

ARIMA ‚Üí AR + MA + differencing (handles trend).

SARIMA ‚Üí ARIMA + seasonality.

SARIMAX ‚Üí SARIMA + external factors.

```python
'''Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components '''
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.datasets import get_rdataset

# Load AirPassengers data
data = get_rdataset("AirPassengers").data
ts = pd.Series(data.value.values,
               index=pd.date_range(start="1949-01", periods=len(data.value), freq="M"))
print(ts.head())
plt.figure(figsize=(10,4))
plt.plot(ts, label="AirPassengers")
plt.title("Monthly AirPassengers Data")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()
# Additive decomposition
decomposition = seasonal_decompose(ts, model="multiplicative")

# Plot components
decomposition.plot()
plt.show()
```

```python
'''Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
# Simulated taxi fares: most between $5‚Äì$50, with some anomalies
np.random.seed(42)
normal_fares = np.random.uniform(5, 50, 500)
anomalies = np.random.uniform(100, 200, 20)  # unusually high fares
fares = np.concatenate([normal_fares, anomalies])

# Create a simple DataFrame
df = pd.DataFrame({"fare": fares})
df["trip_distance"] = np.concatenate([np.random.uniform(1, 20, 500),
                                      np.random.uniform(0.1, 2, 20)])  # short trips but huge fares
# Fit Isolation Forest
iso = IsolationForest(contamination=0.05, random_state=42)
df["anomaly"] = iso.fit_predict(df[["fare", "trip_distance"]])
plt.figure(figsize=(8,6))
colors = np.where(df["anomaly"] == -1, "red", "blue")

plt.scatter(df["trip_distance"], df["fare"], c=colors, alpha=0.6)
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare ($)")
plt.title("Isolation Forest Anomaly Detection (NYC Taxi Fare)")
plt.show()
```

```python
'''Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results. '''
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.datasets import get_rdataset

# Load AirPassengers data
data = get_rdataset("AirPassengers").data
ts = pd.Series(data.value.values,
               index=pd.date_range(start="1949-01", periods=len(data.value), freq="M"))
model = SARIMAX(ts, order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit(disp=False)
forecast = results.get_forecast(steps=12)
pred_ci = forecast.conf_int()
plt.figure(figsize=(10,6))
plt.plot(ts, label="Observed")
plt.plot(forecast.predicted_mean, label="Forecast", color="red")

# Confidence interval
plt.fill_between(pred_ci.index,
                 pred_ci.iloc[:,0],
                 pred_ci.iloc[:,1],
                 color="pink", alpha=0.3)

plt.title("SARIMA Forecast - AirPassengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

```python
'''Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib. '''
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
np.random.seed(42)

# Normal cluster
X_normal = 0.3 * np.random.randn(200, 2)
X_normal = np.r_[X_normal + 2, X_normal - 2]  # two clusters

# Add anomalies
X_outliers = np.random.uniform(low=-6, high=6, size=(20, 2))

# Combine
X = np.r_[X_normal, X_outliers]
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)

# -1 = anomaly, 1 = normal
plt.figure(figsize=(8,6))
colors = np.where(y_pred == -1, "red", "blue")

plt.scatter(X[:,0], X[:,1], c=colors, alpha=0.6)
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Question 10: You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:
‚óè How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?
‚óè Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?
‚óè How would you validate and monitor the performance over time?
‚óè How would this solution help business decisions or operations?

Ans.Step 1: Anomaly Detection in Streaming Data

We need to flag abnormal spikes/drops in near real-time.

Candidate models:

Isolation Forest ‚Üí scalable, works well with high-dimensional features (energy usage + weather).

LOF ‚Üí good for density-based local anomalies, but less scalable for streaming.

DBSCAN ‚Üí better for clustering, not ideal for continuous streaming.

üëâ Choice: Isolation Forest, because it handles large-scale, high-dimensional, real-time streams and can update incrementally.

Example: A sudden 3x jump in usage in one region = anomaly.

We can retrain on sliding windows (e.g., last 24h) to keep it adaptive.

üîπ Step 2: Short-term Forecasting Model

We want every 15 min demand forecast (next few hours).

ARIMA: handles trend, but not seasonality or external factors.

SARIMA: handles seasonality (e.g., daily usage patterns).

SARIMAX: SARIMA + external variables (like weather).

üëâ Choice: SARIMAX, because energy demand depends on temperature, humidity, time of day, region.

Example: Hot day ‚Üí AC usage spikes ‚Üí higher demand forecast.

üîπ Step 3: Validation & Monitoring

Backtesting: Split data into rolling windows (train on past, test on future).

Metrics: RMSE or MAPE for forecast accuracy; Precision/Recall for anomaly detection.

Real-time monitoring:

Compare forecast vs actual ‚Üí large residuals = anomaly flag.

Recalibrate model weekly/monthly as consumption patterns shift.

Drift detection: If error consistently grows, retrain with fresh data.

üîπ Step 4: Business Value

This solution helps the grid operator to:

Balance supply & demand: Predict demand peaks so enough power is generated.

Prevent failures: Detect sudden drops/spikes = potential outages, theft, or equipment faults.

Optimize costs: Avoid overproducing power when demand is low.

Improve customer reliability: Proactively respond before customers experience blackouts.

Stream data ‚Üí detect anomalies with Isolation Forest ‚Üí forecast demand with SARIMAX ‚Üí validate with backtesting & monitoring ‚Üí use insights for grid stability and cost efficiency.
